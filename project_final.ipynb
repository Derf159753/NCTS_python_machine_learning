```python
from keras.models import Model
from keras.models import Sequential
from keras.layers import Dense, Activation, Input, concatenate, Conv2D, MaxPool2D, Flatten
from keras.optimizers import SGD
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
```

```python
X1=[]
for root, dirs, files in os.walk(loc_Para_output, topdown=False):
    for name in files:
        data=cv2.imread(loc_Para_output+os.path.join(name))
        X1.append(data)
        
X2=[]
for root, dirs, files in os.walk(loc_Unin_output, topdown=False):
    for name in files:
        data=cv2.imread(loc_Unin_output+os.path.join(name))
        X2.append(data)
```

把資料讀入

```python
X1=np.array(X1)
print(X1.shape)

X2=np.array(X2)
print(X2.shape)
```

檢查資料形狀，分別是感染和未感染細胞圖片各13779張，長寬各50，有RGB三個色彩

```python
X=np.concatenate((X1,X2),axis=0)

print(X.shape)

data=X[27557,:,:]
plt.imshow(data,plt.cm.gray)
```

合併資料，並讀出其中一張

```python
Y1 = np.ones(X1.shape[0])
Y2 = np.zeros(X2.shape[0])

Y=np.concatenate((Y1,Y2),axis=0)
print(Y.shape)
```

建立圖片的答案，因為目前還是有序的資料，前半是感染細胞設定為1，後半是為感染細胞設定為0

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
```

因為此時的資料是有順序的，所以在切分學習資料和驗證資料時要建立隨機種子。

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=seed)
y_train = np_utils.to_categorical(y_train, 2)
y_test = np_utils.to_categorical(y_test, 2)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

將X和Y資料集進行切分，用其中五分之一作為驗證資料，剩下的是學習資料

```python
index=np.arange(x_train.shape[0])
np.random.shuffle(index)
print(index[0:20])
 
x_train=x_train[index,:]
y_train=y_train[index]
```

```python
import keras
from keras.layers import Dense, Conv2D
from keras.layers import Flatten
from keras.layers import MaxPooling2D, GlobalAveragePooling2D
from keras.layers import Activation
from keras.layers import BatchNormalization
from keras.layers import Dropout
from keras.models import Sequential
from keras import backend as K

from keras import optimizers
```

```python
model2 = Sequential()
```

```python
def CNNbuild(height, width, classes, channels):
    model2 = Sequential()
    
    inputShape = (height, width, channels)
    chanDim = -1
    
    if K.image_data_format() == 'channels_first':
        inputShape = (channels, height, width)
    model2.add(Conv2D(32, (3,3), activation = 'relu', input_shape = inputShape))
    model2.add(MaxPooling2D(2,2))
    model2.add(BatchNormalization(axis = chanDim))
    model2.add(Dropout(0.2))

    model2.add(Conv2D(32, (3,3), activation = 'relu'))
    model2.add(MaxPooling2D(2,2))
    model2.add(BatchNormalization(axis = chanDim))
    model2.add(Dropout(0.2))

    model2.add(Conv2D(32, (3,3), activation = 'relu'))
    model2.add(MaxPooling2D(2,2))
    model2.add(BatchNormalization(axis = chanDim))
    model2.add(Dropout(0.2))

    model2.add(Flatten())
    
    model2.add(Dense(512, activation = 'relu'))
    model2.add(BatchNormalization(axis = chanDim))
    model2.add(Dropout(0.5))
    model2.add(Dense(classes, activation = 'softmax'))
    
    return model2
```

```python
#instantiate the model
height = 50
width = 50
classes = 2
channels = 3
model2 = CNNbuild(height = height, width = width, classes = classes, channels = channels)
model2.summary()
```

```python
model2.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])
```

```python
#fit the model onto the dataset
h = model2.fit(x_train, y_train, epochs = 20, batch_size = 32)
```

```python
score2 = model2.evaluate(x_test, y_test, batch_size=100)
print("Loss: %f" %score2[0])
print("準確率: %f" %(score2[1]*100))
```

可以看到他的模式準確度比我的高多了，其精確度可以達到98%，所以就看看他的模式跟我的有什麼差異，可以從哪邊改善起

模式三：
我發現模式一跟二最大的差別在於每層神經網路之間有多進行Normalization和Dropout
Normalization會將上一層神經網路的輸出項進行標準化，轉換成平均值為0，標準差為1
Dropout則會隨機將部分上一層神經網路的輸出變成0，可以避免過擬合問題
所以就將模式一的其他部份都維持不變，但是在每一層神經網路之間都加入Normalization和Dropout

```python
model3 = Sequential()

model3.add(Conv2D(32, (7, 7), padding='same', input_shape=(50, 50, 3)))
model3.add(Activation('relu'))
model3.add(MaxPool2D(pool_size=(2,2)))
model3.add(BatchNormalization(axis = -1))
model3.add(Dropout(0.2))

model3.add(Conv2D(36, (7, 7), padding='same', input_shape=(50, 50, 3)))
model3.add(Activation('relu'))
model3.add(MaxPool2D(pool_size=(2,2)))
model3.add(BatchNormalization(axis = -1))
model3.add(Dropout(0.2))

model3.add(Conv2D(40, (7, 7), padding='same', input_shape=(50, 50, 3)))
model3.add(Activation('relu'))
model3.add(MaxPool2D(pool_size=(2,2)))
model3.add(BatchNormalization(axis = -1))
model3.add(Dropout(0.2))

model3.add(Flatten())

model3.add(Dense(16))
model3.add(Activation('relu'))
model3.add(BatchNormalization(axis = -1))
model3.add(Dropout(0.2))

model3.add(Dense(2))
model3.add(Activation('softmax'))
```

```python
model3.compile(loss='mse', optimizer=SGD(lr=0.07), metrics=['accuracy'])
model3.summary()
```

```python
model3.fit(x_train, y_train, batch_size=100, epochs=20)
```

```python
score3 = model3.evaluate(x_test, y_test, batch_size=100)
print("Loss: %f" %score3[0])
print("準確率: %f" %(score3[1]*100))
```

其結果顯示加入Normalization和Dropout兩道手續後真的可以大幅改善整個模式，其精確度可以達到近96%
故在此就做為本期末計畫的一個段落
