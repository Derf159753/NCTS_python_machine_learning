```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx?DATA=3B7D6EF7-3636-4E0C-85CC-A5A37B4A90C2')
data=data.drop([0])
data
```

```python
data.loc[data['site_id']=='新北市板橋區']
```

```python
data['birth_total']=pd.to_numeric(data['birth_total'])
data['death_total']=pd.to_numeric(data['death_total'])
data['net']=data['birth_total']-data['death_total']
data
```

```python
plt.hist(data.net)
plt.show()
```
