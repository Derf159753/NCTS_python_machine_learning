```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model as skl
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data=pd.read_csv('NA8101A1Ac.csv')
data=data.drop([0,68,69])
data
```

GDP:國內生產毛額
PPP:國內生產毛額(每人)
GNI:國民所得毛額
IPP:平均每人所得
NI:國民所得
NIPP:國民所得(每人)
rate:年增長率

```python
reg = skl.LinearRegression()
Y=data.population_rate
X=pd.concat([data.GDP_rate,data.GNI_rate,data.IPP_rate,data.NIPP_rate,data.NI_rate,data.PPP_rate],axis=1)
X_train=X[:-10]
X_test=X[-10:]
Y_train=Y[:-10]
Y_test=Y[-10:]
```

```python
reg.fit(X_train,Y_train)
Y_pred=reg.predict(X_test)
```

```python
print('Coefficients: \n', reg.coef_)
print("Mean squared error: %.2f"
      % mean_squared_error(Y_test,Y_pred))
```
