程式目的：擷取水位站資料，並根據擷取的時間把資料分類完成並儲存
資料描述：該水位資料每十分鐘回傳一筆，但是同一時間擷取的資料不一定是同一時間的。
程式描述：採用擷取動作執行的時間進行分類，依照擷取的時間分別存在對應的年月日資料夾中，如果還沒有，就建立對應的資料夾
改進目標：
1.需要建立爬蟲不間斷的每小時進行擷取，可能需要建立伺服器(不確定)
2.同一時間擷取到的資料時間並不一致

```python
import pandas as pd
import os
import csv
import datetime
```

```python
data=pd.read_csv('http://data.wra.gov.tw/Service/OpenData.aspx?format=csv&amp=&id=2D09DB8B-6A1B-485E-88B5-923A462F475C')
data
```

```python
year=datetime.datetime.now().year
month=datetime.datetime.now().month
day=datetime.datetime.now().day
year_path='data/'+str(year)
year_path
```

```python
if os.path.isdir(year_path):
    month_path=year_path+'/'+str(month)
else:
    os.makedirs(year_path)
```

```python
if os.path.isdir(month_path):
    day_path=month_path+'/'+str(day)
else:
    os.makedirs(month_path)
```

```python
if os.path.isdir(day_path):
    ouput_name=day_path+'/'+str(datetime.datetime.now().hour)+'.csv'
else:
    os.makedirs(day_path)
```

```python
data.to_csv(ouput_name)
```
